```python
df.dropna(inplace=True)
```

```python
df['Company Name']=df['Company Name'].replace('\d+', '',regex=True)
df['Company Name']=df['Company Name'].replace('\n.', '',regex=True)
df['Company Name']
```

```python
#Formatiing the Salary part for easy readablity
df['Salary Estimate']=df['Salary Estimate'].apply(lambda x: x.split('(')[0])
df['Salary Estimate']=df['Salary Estimate'].apply(lambda x: x.replace('$','').replace('K',''))
```

```python
df['Salary Estimate']
```

```python
df.sort_values('Salary Estimate')
```

```python
# df['Salary Estimate']=df['Salary Estimate'].replace('\D+', '',regex=True)
# df['Salary Estimate']=df['Salary Estimate'].replace('\W+','',regex=True) 
```

```python
df['Min Salary']=df['Salary Estimate'].apply(lambda x: x.split('-')[0])
#Converting the obj value to string value
df['Min Salary']=df['Min Salary'].astype(int)
df['Min Salary']
```

```python
df['Min Salary'].unique()
```

```python
df['Max Salary']=df['Salary Estimate'].apply(lambda x: x.split('-')[1])
df['Max Salary']=df['Max Salary'].astype(int)
df['Max Salary']
```

```python
df['Max Salary'].unique()
```

```python
df['Avg Salary']=(df['Max Salary']+df['Min Salary'])/2
df['Avg Salary']=df['Avg Salary'].astype(int)
df['Avg Salary']
```

```python
df['Avg Salary'].unique()
```

```python
df['Skill_Python']=df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['Skill_Python'].value_counts()
```

```python
df['skill_Excel']=df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df['skill_Excel'].value_counts()
```

```python
df['Skill_AWS']=df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['Skill_AWS'].value_counts()
```

```python
df['Skill_hadoop']=df['Job Description'].apply(lambda x: 1 if 'hadoop' in x.lower() else 0)
df['Skill_hadoop'].value_counts()
```

```python
df['Skill_Bigdata']=df['Job Description'].apply(lambda x: 1 if 'big data' in x.lower() else 0)
df['Skill_Bigdata'].value_counts()
```

```python
df['Work_State']=df['Location'].str[-2:]
```

```python
df['Work_State']
```

```python
df['Work_State'].value_counts()
```

```python
#Data Cleaning - Converting  Size data to int form
df['Size']=df['Size'].replace('employees','',regex=True)
df['Size']
```

```python
df['Review']=df['Rating'].apply(lambda x: "Great Work Culture" if x>=3.5 else  "Not Recomended for Work")
df['Review']
```

```python
df['Founded']=df['Founded'].apply(lambda x: "Data Unavialable" if x==-1 else x )
df['Rating']=df['Rating'].apply(lambda x: "Data Unavialable" if x==-1 else x)
df['Size']=df['Size'].apply(lambda x: "Data Unavialable" if x=="-1" else x)
df['Industry']=df['Industry'].apply(lambda x: "Data Unavialable" if x=="-1" else x )
df['Competitors']=df['Competitors'].apply(lambda x: "Data Unavialable" if x=="-1" else x )
df['Sector']=df['Sector'].apply(lambda x: "Data Unavialable" if x=="-1" else x )
df['Revenue']=df['Revenue'].apply(lambda x: "Missing" if x=="-1" else x )
df['Headquarters']=df['Headquarters'].apply(lambda x: "Data Unavialable" if x=="-1" else x)
df['Type of ownership']=df['Type of ownership'].apply(lambda x: "Data Unavialable" if x=="-1" else x)
```
